```python
from google.colab import files
uploaded = files.upload()
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('student_data.csv')

# Outlier Removal
numerical_cols = ['age', 'absences', 'G1', 'G2', 'G3']
print("Before outlier removal:\n", df[numerical_cols].describe())

Q1 = df['absences'].quantile(0.25)
Q3 = df['absences'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['absences'] >= Q1 - 1.5 * IQR) & (df['absences'] <= Q3 + 1.5 * IQR)]

print("After outlier removal:\n", df[numerical_cols].describe())

# EDA Visualization: Boxplot of absences
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['absences'])
plt.title('Boxplot of Absences After Outlier Removal')
plt.xlabel('Absences')
plt.savefig('absences_boxplot.png')
from google.colab import files
files.download('absences_boxplot.png')
plt.show()
```
