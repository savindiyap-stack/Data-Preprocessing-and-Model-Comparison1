```python
from google.colab import files
uploaded = files.upload()
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
df = pd.read_csv('student_data.csv')

# Encoding Categorical Variables
categorical_cols = df.select_dtypes(include=['object']).columns
print("Categorical columns before encoding:", categorical_cols)

le = LabelEncoder()
for col in categorical_cols:
    df[col] = le.fit_transform(df[col].astype(str))

print("Data after encoding:\n", df.head())

# EDA Visualization: Count plot of encoded sex
plt.figure(figsize=(8, 6))
sns.countplot(x='sex', data=df)
plt.title('Count of Male (1) vs Female (0) Students After Encoding')
plt.xlabel('Sex (0=F, 1=M)')
plt.ylabel('Count')
plt.savefig('sex_countplot.png')
from google.colab import files
files.download('sex_countplot.png')
plt.show()
```
